```python
import numpy as np
import pandas as pd
```

```python
athlet_df = pd.read_csv(r"sport_athletes.csv")
athlet_df.sample(5)
```

#### Question 1

```python
athlet_df.groupby('Sport Type')[['Height(cm)', 'Weight(kg)', 'Wins']].max().rename(columns={'Weight(kg)': 'max_weight', 'Height(cm)':'max_height', 'Wins':'max_wins'})
```

#### Question 2

```python
athlet_df.groupby('Sport Type')['Weight(kg)'].min().sort_values()
```

#### Question 3

```python
athlet_df.groupby('Sport Type').apply(lambda group: group.nlargest(2, 'Height(cm)'))
```

#### Question 4

```python
athlet_df.groupby(['Sport Type', 'Country']).max()
```

```python
athlet_df.groupby(['Sport Type', 'Country']).max().loc[('Basketball', 'USA')]
```

```python
athlet_df.groupby(['Sport Type', 'Country']).max().xs(level='Sport Type', key='Soccer').reset_index()
```

```python
athlet_df[athlet_df['Sport Type'].isin(['Soccer', 'Tennis'])].groupby(['Sport Type', 'Country']).max()
```

```python
athlet_df.groupby(['Sport Type', 'Country']).max().swaplevel().sort_index()
```

```python
athlet_df.groupby(['Sport Type', 'Country']).agg({'Height(cm)': ['max', 'min', 'mean', 'std'], 'Wins': ['max', 'min', 'mean', 'std']})
```

#### Question 5

```python
new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010'] })

athlet_df = pd.concat([athlet_df, new_rows], axis=0).reset_index()
athlet_df.tail(5)
```

#### Question 6

```python
athlet_df['Olympic Date'] = pd.to_datetime(athlet_df['Olympic Date'], format='%d/%m/%Y', errors='coerce')
athlet_df.set_index('Olympic Date', inplace=True)
```

```python
athlet_df.resample(rule='YE').count()['Wins'].replace(0, np.nan).dropna()
```

#### Question 7

```python
smoker_df = pd.read_csv(r"athletes_smoking.csv")
smoker_df.sample(5)
```

```python
inner = pd.merge(athlet_df, smoker_df, how='inner')
inner.sample(5)
```

```python
outer = pd.merge(athlet_df, smoker_df, how='outer')
outer.sample(5)
```
